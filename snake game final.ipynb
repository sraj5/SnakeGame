```python
import pygame
import random
pygame.init()
width=600
height=400
snake_block=10
dis=pygame.display.set_mode((width,height))
pygame.display.set_caption("My Snake Game")
pygame.display.update()
red=(160,0,0)
black=(0,0,0)
white=(255,255,255)
grey=(150,150,150)
orange=(255,150,20)
fo=(0,200,240) #colour for food
ran=(126,145,190)
dis.fill(white)
snake_speed=20
# game_over=False
# x=300
# y=200
# x_change=0
# y_change=0
# game_close=False   #in this program game close means losing the current game and game over means closing the game window
clock=pygame.time.Clock()
font_style=pygame.font.SysFont("freesans",25)
lost_img=font_style.render("You Lost! Press P to play again or Q to quit",True,red)#the text in "" will be shown with properties
#defined in font_style variable and then converted to image and then shown on the screen
#True is for whether we want sharp edges for our letters or not

def our_snake(snake_block,snake_list): #block is the snake body and list is the coordinates snake has been
    for x in snake_list:
        pygame.draw.rect(dis,red,(x[0],x[1],snake_block,snake_block)) #actually snake_block is twice because its dimension of each rect i.e. 10X10.
def thescore(score):
    value=font_style.render("Your Score:"+str(score),True,fo)#true is for sharp edges of the text
    dis.blit(value,[0,0]) #display image at (0,0) i.e. top left corner
def game_loop():
    game_over=False
    game_close=False
    x=int(width/2)
    y=int(height/2)
    x_change=0
    y_change=0
#     foodx=10*random.randint(0,59) #food can have x coordinates from 0 to 590 and as the size of food will be 10X10
    #the random number generated should be a multiple of 10. So we get a random number between 0 and 59 and multiply by 10
#     foody=10*random.randint(0,39)
    foodx=snake_block*random.randint(0,(width/snake_block-1))
    foody=snake_block*random.randint(0,(height/snake_block-1))
    snake_list=[]
    length_of_snake=1
    while(game_over==False):
        while(game_close==True):#to check whether player wants to close game after losing or restart
            dis.fill(white)
            dis.blit(lost_img,[100,100]) #100,100 is position
            pygame.display.update()
            for event in pygame.event.get():
                if event.type==pygame.KEYDOWN:
                    if event.key==pygame.K_q:
                        game_over=True
                        game_close=False
                    if event.key==pygame.K_p:
                        game_loop()#restart the game_loop function and thus restart the game
                if event.type==pygame.QUIT: ##so that the cross button at top right can still close the game
                    game_over=True
                    game_close=False
        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                game_over=True
            if event.type==pygame.KEYDOWN:
                if event.key==pygame.K_UP:
                    x_change=0
                    y_change=-10
                elif event.key==pygame.K_DOWN:
                    x_change=0
                    y_change=10
                elif event.key==pygame.K_LEFT:
                    x_change=-10
                    y_change=0
                elif event.key==pygame.K_RIGHT:
                    x_change=10
                    y_change=0
        if x>=width or x<0 or y>=height or y<0:  #if it goes beyond the boundary, game over
            game_close=True
        x+=x_change
        y+=y_change
        dis.fill(grey)
        #ygame.draw.rect(dis,red,(x,y,10,10))#draw a rect for the snake at (x,y) as top left corner of rect and dimension of
        #rect is 10X10
        pygame.draw.rect(dis,fo,(foodx,foody,snake_block,snake_block))
        snake_head=[]
        snake_head.append(x)
        snake_head.append(y)
        snake_list.append(snake_head) #so the head position of snake will always be the last and thus -1 position
        if len(snake_list)>length_of_snake:
            del snake_list[0]
        for i in snake_list[:-1]: #its a function of game ended if the snake eats itself
            #from first till just before the last position as last is head
            if i==snake_head:
                game_close=True
        our_snake(10,snake_list)
        if x==foodx and y==foody:
            length_of_snake+=1
            foodx=snake_block*random.randint(0,(width/snake_block-1))
            foody=snake_block*random.randint(0,(height/snake_block-1))
        thescore(length_of_snake-1)
        pygame.display.update()
        clock.tick(snake_speed)   #in frame per second speed

    #     pygame.display.update()
    pygame.quit()
game_loop()      #calling the function for first time
```
